**Importing Essential Libraries for Image Classification with Keras.**

---



```python
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


from sklearn.model_selection import train_test_split
from sklearn.utils import shuffle
from keras.models import Sequential
from keras.layers import Dense, Flatten, Conv2D, MaxPool2D, Dropout
from keras.optimizers import SGD, Adam
from keras.callbacks import ReduceLROnPlateau, EarlyStopping
from keras.utils import to_categorical
```

**Loading Handwritten Alphabets Dataset from CSV File.**

---



```python
data = pd.read_csv(r"/content/A_Z Handwritten Data/A_Z Handwritten Data.csv").astype('float32')
```

**Preparing Handwritten Alphabets Dataset for Training and Testing.**

---



```python
X = data.drop('0',axis = 1)
y = data['0']


train_x, test_x, train_y, test_y = train_test_split(X, y, test_size = 0.2)

train_x = np.reshape(train_x.values, (train_x.shape[0], 28,28))
test_x = np.reshape(test_x.values, (test_x.shape[0], 28,28))


word_dict = {0:'A',1:'B',2:'C',3:'D',4:'E',5:'F',6:'G',7:'H',8:'I',9:'J',10:'K',11:'L',12:'M',13:'N',14:'O',15:'P',16:'Q',17:'R',18:'S',19:'T',20:'U',21:'V',22:'W',23:'X', 24:'Y',25:'Z'}
```

**Visualizing the Distribution of Handwritten Alphabets in the Dataset** :


---


This code counts the occurrences of each alphabet in the dataset and visualizes the distribution using a horizontal bar chart. It helps in understanding the dataset's balance across different letters, ensuring a diverse representation for training a machine learning model.

```python
y_int = np.int0(y)
count = np.zeros(26, dtype='int')
for i in y_int:
    count[i] +=1

alphabets = []
for i in word_dict.values():
    alphabets.append(i)


fig, ax = plt.subplots(1,1, figsize=(10,10))
ax.barh(alphabets, count)

# naming the x axis
plt.xlabel("Number of elements")
# naming the y axis
plt.ylabel("Alphabets")
# giving a title
plt.title("Plotting the number of alphabets")
# Turn on the minor TICKS, which are required for the minor GRID
plt.minorticks_on()
# Customize the major grid
plt.grid(which='major', linestyle='-', linewidth='0.5', color='red')
# Customize the minor grid
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')

plt.show()
```

**Displaying Shuffled Handwritten Alphabets:**


---


This code shuffles a subset of the training data and displays a 3x3 grid of handwritten alphabet images. The images are thresholded to enhance visibility, providing a visual representation of the shuffled dataset.

```python
shuff = shuffle(train_x[:100])
fig, ax = plt.subplots(3,3, figsize = (10,10))
axes = ax.flatten()
for i in range(9):
    _, shu = cv2.threshold(shuff[i], 30, 200, cv2.THRESH_BINARY)
    axes[i].imshow(np.reshape(shuff[i], (28,28)), cmap=plt.get_cmap('gray'))
plt.show()
```

**Reshaping Data for Model Creation:**


---


This code reshapes the training and testing data to be suitable for a convolutional neural network (CNN). It converts the data to a 4D array, representing the image dimensions with an additional channel (1 for grayscale). Additionally, it one-hot encodes the alphabet labels for both training and testing datasets, preparing them for model training.

```python
# Reshape data for model creation
train_X = train_x.reshape(train_x.shape[0],train_x.shape[1],train_x.shape[2],1)
print("The new shape of train data: ", train_X.shape)

test_X = test_x.reshape(test_x.shape[0], test_x.shape[1], test_x.shape[2],1)
print("The new shape of train data: ", test_X.shape)

train_yOHE = to_categorical(train_y, num_classes = 26, dtype='int')
print("The new shape of train labels: ", train_yOHE.shape)

test_yOHE = to_categorical(test_y, num_classes = 26, dtype='int')
print("The new shape of test labels: ", test_yOHE.shape)
```

**Defining Convolutional Neural Network (CNN) Model:**


---


This code defines a sequential CNN model with convolutional and pooling layers, followed by dense layers. The model is designed for handwritten alphabet recognition with a softmax activation function for the output layer having 26 classes.

```python
model = Sequential()

model.add(Conv2D(filters=32, kernel_size=(3, 3), activation='relu', input_shape=(28,28,1)))
model.add(MaxPool2D(pool_size=(2, 2), strides=2))

model.add(Conv2D(filters=64, kernel_size=(3, 3), activation='relu', padding = 'same'))
model.add(MaxPool2D(pool_size=(2, 2), strides=2))

model.add(Conv2D(filters=128, kernel_size=(3, 3), activation='relu', padding = 'valid'))
model.add(MaxPool2D(pool_size=(2, 2), strides=2))

model.add(Flatten())

model.add(Dense(64,activation ="relu"))
model.add(Dense(128,activation ="relu"))

model.add(Dense(26,activation ="softmax"))
```

**Compiling and Training the CNN Model:**


---


This code compiles the CNN model using the Adam optimizer, categorical crossentropy loss, and accuracy as the evaluation metric. It then trains the model for one epoch on the training data, validating on the testing data, and records the training history.

```python
model.compile(optimizer = Adam(learning_rate=0.001), loss='categorical_crossentropy', metrics=['accuracy'])

history = model.fit(train_X, train_yOHE, epochs=1, validation_data = (test_X,test_yOHE))
```

**Model Summary, Saving, and Training History:**


---


This code displays the summary of the trained CNN model, saves the model as "model_hand.h5," and prints the validation and training accuracy as well as the validation and training loss obtained during training.

```python
model.summary()
model.save(r'model_hand.h5')

print("The validation accuracy is :", history.history['val_accuracy'])
print("The training accuracy is :", history.history['accuracy'])
print("The validation loss is :", history.history['val_loss'])
print("The training loss is :", history.history['loss'])
```

**Making Predictions on Test Data:**


---


This code visualizes predictions on a subset of the test data. It displays a 3x3 grid of handwritten alphabet images from the test set along with their corresponding predictions made by the trained model.

```python
# Prediction on test data
fig, axes = plt.subplots(3,3, figsize=(8,9))
axes = axes.flatten()

for i,ax in enumerate(axes):
    img = np.reshape(test_X[i], (28,28))
    ax.imshow(img,  cmap=plt.get_cmap('gray'))

    pred = word_dict[np.argmax(test_yOHE[i])]
    ax.set_title("Prediction: "+pred)
```
